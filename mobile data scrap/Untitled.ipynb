```python
import requests
import bs4
import pandas as pd
import numpy as np
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
src=requests.get('https://pricebaba.com/search?q=j7+max')
```

```python
soup=bs4.BeautifulSoup(src.text, 'html.parser')
```

```python
#print(soup.prettify())
```

```python
box=soup.find('div', class_='stack-inline m-b-xl flex-grid prdts-cnt')
```

```python
cards=box.find_all('div', class_='col-3 p-h-xs v-al-top flt-l' )
```

```python
len(cards)
```

```python
urls=[]
name=[]
price=[]
details=[]
for card in cards:
    d=[]
    sub1=card.find('div', class_='col-12 v-al-mdl p-h-xs')
    a_tag=sub1.find('a')
    urls.append(str(a_tag.attrs['href']))
    name.append(a_tag.text.strip())
    subsub1=card.find('div',class_='d-block')
    price.append(subsub1.find('span').text.strip())
    sub2=card.find('div', class_='prdp-spc p-v-xs')
    li=sub2.find_all('li')
    for i in li:
        d.append(i.find('span').text.strip())
    details.append(d)
```

```python
screen=[]
camera=[]
ram=[]
battery=[]
for i in details:
    try:
        screen.append(i[0])
    except:
        screen.append(np.nan)
    try:
        camera.append(i[1])
    except:
        camera.append(np.nan)
    try:
        ram.append(i[2])
    except:
        ram.append(np.nan)
    try:
        battery.append(i[3])
    except:
        battery.append(np.nan)
```

```python
d={'name':name,'price':price,'url':urls,'screen':screen,'camera':camera,'ram':ram,'battery':battery}
df=pd.DataFrame(d)
```
