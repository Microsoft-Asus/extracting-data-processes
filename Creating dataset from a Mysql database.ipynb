```python
query='SELECT fname,lname, task_id, score,Late sub,winner FROM evaluation e JOIN users u ON u.user_id=e.user_id'

mycursor.execute(query)

data=mycursor.fetchall()
```

```python
for i in data:
    print(i)
```

```python
import pandas as pd
```

```python
name=[]
task_id=[]
score=[]
late_sub=[]
winner=[]

for i in data:
    name.append(i[0]+' '+i[1])
    task_id.append(i[2])
    score.append(i[3])
    late_sub.append(i[4])
    winner.append(i[5])
    
    d={'Name':name, 'Task Id':task_id,'Score':score,'late sub':late_sub,'winner':winner}
    
    df=pd.DataFrame(d)
```
