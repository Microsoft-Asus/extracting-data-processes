```python
import pandas as pd
import numpy as np
import requests
import bs4
```

```python
src=requests.get('https://www.mysmartprice.com/computer/pricelist/processors-price-list-in-india.html')
soup=bs4.BeautifulSoup(src.text, 'html.parser')
```

```python
boxes=soup.find_all('div', class_='prdct-item__dtls')
prices=[]
model=[]
for i in boxes:
    try:
        var=i.find('div', class_='prdct-item__prc main-wrpr__cols1')
        prices.append(var.find_all('span'))
    except:
        prices.append(np.nan)
    try:
        model.append(i.find('a').text.strip())
    except:
        model.append(np.nan)
        
price=[]
for i in range(len(prices)):
    try:
        price.append(prices[i][1].text)
    except:
        price.append(np.nan)
    
```

```python
len(model)
```

```python
len(price)
```

```python
d={'model':model,'prices':price}
df1=pd.DataFrame(d)
```
